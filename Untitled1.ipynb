# Trigonometrische Funktionen
Die Graphen der Sinus- und Kosinusfunktion k√∂nnen auf verschiedene Weise ver√§ndert werden. Sie k√∂nnen in ùë•‚àí und ùë¶‚àí Richtung verschoben, gestreckt oder gestaucht sein.

Eine ver√§nderte trigonometrische Funktion kann zum Beispiel so aussehen: ùëì(ùë•)=3‚ãÖùë†ùëñùëõ(2‚ãÖùë•+1)‚àí1.

Um die Ver√§nderungen leichter beschreiben zu k√∂nnen, klammert man den Faktor vor dem ùë• aus: ùëì(ùë•)=3‚ãÖùë†ùëñùëõ(2‚ãÖ(ùë•+12))‚àí1.

Allgemeine Form
Sinus: f(x) = a ‚ãÖ sin (b ‚ãÖ (x + c)) + d
Kosinus: f(x) = a ‚ãÖ cos (b ‚ãÖ (x + c)) + d
Tangent: f(x) = a ‚ãÖ tan(b ‚ãÖ (x + c)) + d
Die reellen Parameter ùëé,ùëè,ùëê,ùëë bestimmen, wie der Graph genau ver√§ndert wird.

Bemerkung: Nicht nur trigonometrische Funktionen lassen sich so ver√§ndern. Unter den folgenden Links findest du wie man den Graphen einer beliebigen Funktion verschiebt oder staucht, oder streckt.

Einfluss der Parameter auf den Funktionsgraphen
Beobachtung an Beispielen

### 1. Betrachte f(x) = sin (2 ‚ãÖ x) + 1.

‚áíùëë  ver√§ndert also die Ruhelage der Funktion.

2. Danach wird b vom Startwert  1  beginnend bis zum Endwert  2  ver√§ndert. W√§hrenddessen staucht sich der Funktionsgraph in  ùë•‚àíùëÖùëñùëê‚Ñéùë°ùë¢ùëõùëî  zusammen; die Wellenberge und Wellent√§ler r√ºcken enger aneinander, die Periode der Funktion wird kleiner. Beim Endwert  ùëè=2  ist die Periode nur noch  œÄ  statt  2œÄ .  ‚áíùëè  ver√§ndert also die Periode der Funktion.

### 2. Betrachte g(x) = 2 ‚ãÖ cos(x ‚àí 1).

```python
x = np.arange(0,3*np.pi,0.1)   # start,stop,step
y = np.cos(x-1)*2
z = np.cos(x)

plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 2œÄ')  # string must be enclosed with quotes '  '
plt.ylabel('2*cos(x-1) and cos(x)')
plt.title('Plot of cos(x) and 2cos(x-1) from 0 to 2œÄ')
plt.legend(['2cos(x-1)', 'cos(x)'])      # legend entries as seperate strings in a list
plt.show()
```

Der Funktionsgraph ver√§ndert sisch unter dem Einfluss der Parameter a und c:

1.   Zuerst wird  ùëê  vom Startwert  0  beginnend auf den Wert  ‚àí1  ver√§ndert. Dabei verschiebt sich der Funktionsgraph in  ùë•‚àíùëÖùëñùëê‚Ñéùë°ùë¢ùëõùëî  um den Wert  1  nach rechts.

‚áí  ùëê  ver√§ndert also die Lage des Funktionsgraphen in  ùë•‚àíùëÖùëñùëê‚Ñéùë°ùë¢ùëõùëî .

2.   Danach wird  ùëé  vom Startwert  1  beginnend bis zum Endwert  2  ver√§ndert. Dabei wird der Funktionsgraph in  ùë¶‚àíùëÖùëñùëê‚Ñéùë°ùë¢ùëõùëî  gestreckt.

‚áí  ùëé  ver√§ndert also die Amplitude der Funktion.

### √úberblick √ºber den Einfluss der Parameter

### Parameter a

Der Parameter a beeinflusst die Amplitude. Er streckt/staucht den Graph in y-Richtung.

Der Graph hat die Amplitude |a|
a < 0: Der Graph wird zus√§tzlich an der Ruhelage gespiegelt.

```python
x = np.arange(0,3*np.pi,0.1)   # start,stop,step
y = np.sin(x)/(-2)
z = np.sin(x)
p = np.sin(x)*3

plt.plot(x,y,x,z,x,p)
plt.xlabel('x values from 0 to 2œÄ')  # string must be enclosed with quotes '  '
plt.ylabel('- 0,5 * sin(x), sin(x) and 3 * sin(x)')
plt.title('Plot of - 0,5 * sin(x), sin(x) and 3 * sin(x) from 0 to 2œÄ')
plt.legend(['- 0,5*sin(x)', 'sin(x)','3*sin(x)'])      # legend entries as seperate strings in a list
plt.show()
```

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

def plot_func (a):
   # x = np.linspace(0, 2*np.pi,)
    x = np.arange(0,3*np.pi,0.05)
    y = np.sin(x)/(-a)
    z = np.sin(x)
    p = np.sin(x)*a
    
    plt.plot(x, y,x,z,x,p)
    plt.grid()
    plt.xlabel('x values from 0 to 2œÄ')  # string must be enclosed with quotes '  '
    plt.ylabel('- a * sin(x), sin(x) and a * sin(x)')
    plt.title('Plot of - a * sin(x), sin(x) and a * sin(x) from 0 to 2œÄ')
    plt.legend(['- a*sin(x)', 'sin(x)','a*sin(x)'])      # legend entries as seperate strings in a list

interact(plot_func, a = widgets.FloatSlider(value=7.5,min=0.1, max=10.0, step=0.5))


```

### Parameter b

Der Parameter b beeinflusst die Periode. Er streckt/staucht den Graph in  ùë•‚àíùëÖùëñùëê‚Ñéùë°ùë¢ùëõùëî .

Der Graph hat die Periode p = 2œÄ / |b|

b<0: Der Graph wird zus√§tzlich an der senkrechten Achse x = ‚àí c gespiegelt

```python
x = np.arange(0,3*np.pi,0.1)   # start,stop,step
y = np.sin(-2*x)
z = np.sin(x)
p = np.sin(2*x)

plt.plot(x,y,x,z,x,p)
plt.xlabel('x values from 0 to 2œÄ')  # string must be enclosed with quotes '  '
plt.ylabel('sin(-2x), sin(x) and sin(2x)')
plt.title('Plot of sin(-2x), sin(x) and sin(2x) from 0 to 2œÄ')
plt.legend(['sin(-2x)', 'sin(x)','sin(2x)'])      # legend entries as seperate strings in a list
plt.show()
```

```python
def plot_func(b):
   # x = np.linspace(0, 2*np.pi,)
    x = np.arange(0,3*np.pi,0.05)
    y = np.sin(-b*x)
    z = np.sin(x)
    p = np.sin(b*x)
    
    plt.plot(x, y,x,z,x,p)
    plt.grid()
    plt.xlabel('x values from 0 to 2œÄ')  # string must be enclosed with quotes '  '
    plt.ylabel('sin(-bx), sin(x) and sin(bx)')
    plt.title('Plot of sin(-bx), sin(x) and sin(bx) from 0 to 2œÄ')
    plt.legend(['sin(-bx)', 'sin(x)','sin(bx)'])      # legend entries as seperate strings in a list

interact(plot_func, b = widgets.FloatSlider(value=7.5,min=0.01, max=2.0, step=0.1))
```

### Parameter c
Der Parameter c verursacht eine Verschiebung in  ùë•‚àíùëÖùëñùëê‚Ñéùë°ùë¢ùëõùëî 

c > 0 : Verschiebung um c nach links

c < 0 : Verschiebung um c nach rechts

```python
x = np.arange(0,3*np.pi,0.1)   # start,stop,step
y = np.sin(x-1)
z = np.sin(x)
p = np.sin(x+1)

plt.plot(x,y,x,z,x,p)
plt.xlabel('x values from 0 to 2œÄ')  # string must be enclosed with quotes '  '
plt.ylabel('sin(x-1), sin(x) and sin(x+1)')
plt.title('Plot of sin(x-1), sin(x) and sin(x+1) from 0 to 2œÄ')
plt.legend(['sin(x-1)', 'sin(x)','sin(x+1)'])      # legend entries as seperate strings in a list
plt.show()
```

```python
def plot_func(c):
   # x = np.linspace(0, 2*np.pi,)
    x = np.arange(0,3*np.pi,0.05)
    y = np.sin(x-c)
    z = np.sin(x)
    p = np.sin(x+c)
    
    plt.plot(x, y,x,z,x,p)
    plt.grid()
    plt.xlabel('x values from 0 to 2œÄ')  # string must be enclosed with quotes '  '
    plt.ylabel('sin(x-c), sin(x) and sin(x+c)')
    plt.title('Plot of sin(x-c), sin(x) and sin(x+c) from 0 to 2œÄ')
    plt.legend(['sin(x-c)', 'sin(x)','sin(x+c)'])      # legend entries as seperate strings in a list

interact(plot_func, c = widgets.FloatSlider(value=7.5,min=0.1, max=10.0, step=0.5))
```

### Parameter d
Der Parameter  ùëë  beeinflusst die Ruhelage. Er verschiebt den Graph in  ùë¶‚àíùëÖùëñùëê‚Ñéùë°ùë¢ùëõùëî 

d > 0: Verschiebung um d nach oben

d < 0: Verschiebung um d nach unten

Der Graph hat die Ruhelage bei y = d

```python
x = np.arange(0,3*np.pi,0.1)   # start,stop,step
y = np.sin(x)-1
z = np.sin(x)
p = np.sin(x)+2

plt.plot(x,y,x,z,x,p)
plt.xlabel('x values from 0 to 2œÄ')  # string must be enclosed with quotes '  '
plt.ylabel('sin(x)-1, sin(x) and sin(x)+2')
plt.title('Plot of sin(x)-1, sin(x) and sin(x)+2 from 0 to 2œÄ')
plt.legend(['sin(x)-1', 'sin(x)','sin(x)+2'])      # legend entries as seperate strings in a list
plt.show()
```

```python
def plot_func(d):
   # x = np.linspace(0, 2*np.pi,)
    x = np.arange(0,3*np.pi,0.05)
    y = np.sin(x)-d
    z = np.sin(x)
    p = np.sin(x)+d
    
    plt.plot(x, y,x,z,x,p)
    plt.grid()
    plt.xlabel('x values from 0 to 2œÄ')  # string must be enclosed with quotes '  '
    plt.ylabel('sin(x)-d, sin(x) and sin(x)+d')
    plt.title('Plot of sin(x)-d, sin(x) and sin(x)+d from 0 to 2œÄ')
    plt.legend(['sin(x)-d', 'sin(x)','sin(x)+d'])      # legend entries as seperate strings in a list

interact(plot_func, d = widgets.FloatSlider(value=7.5,min=0.01, max=10.0, step=0.5))
```

```python
def plot_func(a,b,c,d):
   # x = np.linspace(0, 2*np.pi,)
    x = np.arange(-0,1*np.pi,0.05)
    y = np.sin(b*(x-c))*a-d
    z = np.sin(x)
    p = np.sin(b*(x+c))*a+d
    
    plt.plot(x, y,x,z,x,p)
    plt.grid()
    plt.xlabel('x values from 0 to 2œÄ')  # string must be enclosed with quotes '  '
    plt.ylabel('a*sin(b(x-c)) - d, sin(x) and a*sin(b(x+c)) + d')
    plt.title('Plot of a*sin(b(x-c)) - d, sin(x) and a*sin(b(x+c)) + d from 0 to 2œÄ')
    plt.legend(['a*sin(b(x-c)) - d', 'sin(x)','asin(b(x+c)) + d'])      # legend entries as seperate strings in a list

interact(plot_func, a = widgets.FloatSlider(value=7.5,min=0.1, max=10.0, step=0.5),
        b = widgets.FloatSlider(value=7.5,min=1, max=5.0, step=0.5),
        c = widgets.FloatSlider(value=7.5,min=1, max=5.0, step=1),
        d = widgets.FloatSlider(value=7.5,min=0.01, max=5.0, step=0.5),)
```
